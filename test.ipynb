```
import numpy as np
import pandas as pd
```

```
base_url = 'https://raw.githubusercontent.com/ashekhar20/league_predict/master/data/'
data_dict=base_url+'data_dictionary.xlsx'
train_file=base_url+'train.csv'
test_file=base_url+'test.csv'
```

```
datadict=pd.read_excel(data_dict)
print(datadict.shape)
datadict.head()
```

```
datadict=datadict.set_index('Variable Name')
datadict.head()
```

```
df1 = pd.read_csv(train_file)
df2 = pd.read_csv(test_file)
df1.shape, df2.shape
```

```
cols=df1.columns
print(len(cols))
cols
```

```
df1['DateTimeOfGame']=pd.to_datetime(df1['DateOfGame']+'T'+df1['TimeOfGame'])
df1['YearOfGame']=df1['DateTimeOfGame'].dt.year
df1['HourOfGame']=df1['DateTimeOfGame'].dt.hour
df1['isTeam1Home'] = df1.apply(lambda x: x['City'] in x['Team 1'], axis=1)
df1.head()
```

```
tmp1=df1[['Team 1','YearOfGame','DateTimeOfGame']]
tmp1.rename(columns={"Team 1": "Team"}, inplace=True)
tmp2=df1[['Team 2','YearOfGame','DateTimeOfGame']]
tmp2.rename(columns={"Team 2": "Team"}, inplace=True)
tmp=pd.concat([tmp1,tmp2])
tt=tmp.groupby(['Team','YearOfGame'])["DateTimeOfGame"].rank("dense", ascending=True)
tmp['TeamMatchNum']=tt.values.astype(int)
print(tmp.shape)
tmp.head(10)
```

```
tmp2=tmp.sort_values(by=['Team','DateTimeOfGame'])
tmp2.head(20)
```

```
tt=df1.groupby(['Team 1','YearOfGame'])["DateTimeOfGame"].rank("dense", ascending=True)
df1['Team1MatchNum']=tt.values.astype(int)
tt=df1.groupby(['Team 2','YearOfGame'])["DateTimeOfGame"].rank("dense", ascending=True)
df1['Team2MatchNum']=tt.values.astype(int)
```
